## What does RT generally cover?

235 RT Twitter Posts are selected for this analysis. Those posts are categorized into eleven groups: 'US Foreign Policy', 'Viral Content', 'Criticizing Democracies', 'Russian Politics', 'US Domestic Politics', 'Championing the Global South', 'Syria', 'Criticizing Israel', 'RT Credibility' and 'Miscellaneous'.  

```python
import pandas as pd
rtdata = pd.read_excel('RTrawdata.xlsx')
categories = rtdata.category.unique()
amount = []
for item in categories:
    amount.append(len(rtdata.where(rtdata.category == item).dropna()))
Categorization = pd.Series(amount, index = categories)
pd.DataFrame(Categorization, columns = ['COUNT'])
```

## Pie Chart for the distribution of RT coverage

```python
import matplotlib.pyplot as plt
Categorization = Categorization.sort_values()
fig, ax = plt.subplots(figsize=(10, 10))
labels = Categorization.index
sizes = list(Categorization)
patches, texts = ax.pie(sizes, startangle=90)
ax.legend(patches, labels, loc="upper right", prop={"size": 12}, bbox_to_anchor=(-0.1, 1))
ax.axis('equal')
plt.tight_layout()
plt.margins(-0.15, -0.15)
plt.show()
```
